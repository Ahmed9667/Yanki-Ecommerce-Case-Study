### Data leaning and Transformation

```python
# Drop misiing values
df = df.dropna(subset=['Order_ID', 'Customer_ID'])
```

```python
# Check for null values
df.info()
```

```python
df.columns
```

```python
# create table customers
customers = df[['Customer_ID', 'Customer_Name','Email', 'Phone_Number']].copy().drop_duplicates().reset_index(drop=True)
customers.head()
```

```python
customers['Customer_ID'].duplicated().any()
```

- Customer_ID has unique values as primary key

```python
# create table products
products = df[['Product_ID',
       'Product_Name', 'Brand', 'Category', 'Price']].copy().drop_duplicates().reset_index(drop =True)
products.head()
```

```python
products['Product_ID'].duplicated().any()
```

- Product_ID is unique as primary key

```python
# create table shipping_address
shipping_address = df[['Customer_ID','Shipping_Address', 'City', 'State', 'Country',
       'Postal_Code']].copy().drop_duplicates().reset_index(drop = True)


shipping_address['Shipping_ID'] = range(1 , len(shipping_address)+1)
# move column Shipping_ID to the first column
shipping_address = shipping_address[['Shipping_ID'] + [x for x in shipping_address.columns if x!= 'Shipping_ID']]


shipping_address.head()
```

```python
shipping_address.shape
```

```python
# create table order payments
payments = df[['Order_ID','Payment_Method','Transaction_Status']].copy().drop_duplicates().reset_index(drop=True)
payments.head()
```

```python
# create table orders
orders  = df[['Order_ID', 'Customer_ID','Product_ID','Quantity', 'Total_Price',
       'Order_Date']].copy().drop_duplicates(subset='Order_ID').reset_index(drop = True)

orders.head()
```

```python
orders['Order_ID'].duplicated().any()
```

- Column Order_ID has unique values as primary key

```python
# Save table to csv files
customers.to_csv(r'Clean Tables\customers.csv' , index = False)
products.to_csv(r'Clean Tables\products.csv' , index = False)
payments.to_csv(r'Clean Tables\payments.csv' , index = False)
shipping_address.to_csv(r'Clean Tables\shipping_address.csv' , index = False)
orders.to_csv(r'Clean Tables\orders.csv' , index = False)
```

```python
# Connect to the new created database Yanki_Ecommerce
def db_connected():
    connection = psycopg2.connect(user = 'postgres', 
                                  host= 'localhost',
                                  password = 'ahly9667',
                                  port = 5432,
                                  database ='yanki_ecommerce')

    return connection

conn = db_connected()
```

```python
def create_tales():
    conn = db_connected()
    cursor = conn.cursor()
    query = """ 
                create schema yanki;
                drop table if exists customers cascade;
                drop table if exists products cascade;
                drop table if exists shipping_address cascade;
                drop table if exists orders cascade;
                drop table if exists payments cascade;
                

                create table if not exists yanki.customers(
                Customer_ID varchar(255) unique primary key ,
                Customer_Name varchar(255),
                Email varchar(255),
                Phone_Number varchar(255)
                );

                create table if not exists yanki.products(
                Product_ID varchar(255) unique primary key ,
                Product_Name varchar(255),
                Brand varchar(255),
                Category varchar(255),
                Price decimal(20,2)
                );

                create table if not exists yanki.shipping_address(
                Shipping_ID int unique primary key,
                Customer_ID varchar(255),
                Shipping_Address varchar(255),
                City varchar(255),
                State varchar(255),
                Country varchar(255),
                Postal_Code  int,
                foreign key (Customer_ID) references yanki.customers(Customer_ID)
                );

                
                create table if not exists yanki.orders(
                Order_ID varchar(255) unique primary key,
                Customer_ID  varchar(255),
                Product_ID varchar(255),
                Quantity int,
                Total_Price decimal(20,2),
                Order_Date date,
                foreign key (Customer_ID) references yanki.customers(Customer_ID) ,
                foreign key (Product_ID) references yanki.products(Product_ID)
                );

                create table if not exists yanki.payments(
                Order_ID varchar(255),
                Payment_Method varchar(255),
                Transaction_Status varchar(255),
                foreign key (Order_ID) references yanki.orders(Order_ID)
                );
            
    """
    cursor.execute(query)
    conn.commit()
    cursor.close()
    conn.close()

create_tales()
```

Insert Records in Tables

```python
# insert records in customer table
import csv
def load_records(csv_path):
    conn = db_connected()
    cursor = conn.cursor()
    with open(csv_path, 'r') as file :
        reader = csv.reader(file)
        next(reader)
        for row in reader:
            try:
                cursor.execute(""" insert into yanki.customers(Customer_ID,Customer_Name,Email,Phone_Number) values(%s,%s,%s,%s); """ , row)
                
            except psycopg2.IntegrityError:
                conn.rollback()
            else:
                conn.commit()

    conn.commit()
    cursor.close()
    conn.close()
            
load_records(r'Clean Tables\customers.csv')                
print('Records added successfully')           
```
